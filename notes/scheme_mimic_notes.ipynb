```python
import sqlite3
```

```python
import pandas as pd

df = pd.read_csv(Path(note_path).joinpath("discharge.csv.gz"), compression='gzip', header=0, sep=',', quotechar='"')
```

```python
con = sqlite3.connect("./../data/medicalnote.db")
df.to_sql("discharge", con, if_exists='replace', index=False)
```
