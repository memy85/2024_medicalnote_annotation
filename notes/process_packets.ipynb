```python
import pandas as pd
import pickle 

from utils import *
config = load_config()

DATA_PATH = config.project_path.joinpath("data/raw/")
```

```python

def normalizer(data) :
    data = data.rename(columns = {"File N" : "file_name",
                           "File Name" : "file_name",
                           "Phrase" : "phrase",
                           "Concept" : "phrase",
                           "Packet N" : "packet",
                           "packet N" : "packet",
                           "Packet Name" : "packet",
                           "Color" : "color",
                           "Color (Y/G)" : "color",
                           "Ranking" : "ranking"
                          })
    columns = data.columns.tolist()
    for idx, col in enumerate(columns) : 
        if "Unnamed" in col : 
            data.drop(col, axis=1, inplace=True)
    
    data['file_name'] = data['file_name'].apply(lambda x : x + ".txt" if '.txt' not in x else x)

    data = data.dropna()

    data['color'] = data['color'].str.lower()
    data['color'] = data['color'].replace({'yellow' : 'y', "green" : 'g'})

    data['ranking'] = data['ranking'].astype("float32")

    return data
```

```python
a3 = pd.read_csv(DATA_PATH.joinpath("packet_3A.txt"), sep='\t')
b3 = pd.read_csv(DATA_PATH.joinpath("packet_3B.txt"), sep='\t')
c3 = pd.read_csv(DATA_PATH.joinpath("packet_3C.txt"), sep='\t')

a3 = normalizer(a3)
b3 = normalizer(b3)
c3 = normalizer(c3)

threes = [a3, b3, c3]
```

```python
a4 = pd.read_csv(DATA_PATH.joinpath("packet_4A.txt"), sep='\t')
b4 = pd.read_csv(DATA_PATH.joinpath("packet_4B.txt"), sep='\t')
c4 = pd.read_csv(DATA_PATH.joinpath("packet_4C.txt"), sep='\t')

a4 = normalizer(a4)
b4 = normalizer(b4)
c4 = normalizer(c4)

fours = [a4, b4, c4]
```

```python
b5 = pd.read_csv(DATA_PATH.joinpath("packet_5B.txt"), sep='\t')
c5 = pd.read_csv(DATA_PATH.joinpath("packet_5C.csv"), sep='\t', encoding='utf-16')

b5 = normalizer(b5)
c5 = normalizer(c5)

fives = [b5, c5]
```
