```python
import json
from utils import *

config = load_config()
PROJECT_PATH = config.project_path
DATA_PATH = PROJECT_PATH.joinpath("data/processed")
```

```python
def load_discharge(n) :
    dataPath = DATA_PATH.joinpath(f"discharge_dataset_augmented_{n}.json")
    with open(dataPath) as f :
        data = json.load(f)
    return data
```

```python
def load_modified(n) :
    with open(PROJECT_PATH.joinpath(f"prompts/modified_finetune_instruction_top{n}.txt"), 'r') as f :
        instruction = f.read()
    return instruction

def change_instruction(case) :
    instruction = case["instruction"]
    if "Extract top 3" in instruction :
        instruction = load_modified(3)
    elif "Extract top 5" in instruction :
        instruction = load_modified(5)
    else :
        instruction = load_modified(10)

    case["instruction"] = instruction
    return case
```

```python
# generating modified prompts
for x in [10,100,1000] :
    df = load_discharge(x)
    modfied_datas = []
    for data in df :
        modified_data = change_instruction(data)
        modfied_datas.append(modified_data)
    with open(DATA_PATH.joinpath(f"modified_discharge_dataset_augmented_{x}.json"), "w") as f :
        json.dump(modfied_datas, f)
    
```
