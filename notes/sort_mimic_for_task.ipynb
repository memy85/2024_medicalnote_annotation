```python
import sqlite3
import pandas as pd
from pathlib import Path

projectPath = Path().cwd().absolute().parents[0]
dataPath = projectPath.joinpath("data/raw")
```

```python
con = sqlite3.connect(dataPath.joinpath('medicalnote.db'))
```

```python

df = pd.read_sql("select * from discharge", con)
```
